```python
from sklearn.preprocessing import LabelEncoder

def label_encoder(column):
    le = LabelEncoder().fit(column)
    print(column.name, le.classes_)
    return le.transform(column)
```

```python
cols = []
for col in df.columns:
    if df[col].dtype == "object":
        cols.append(col)
```

```python
for col in cols:
    df[col] = label_encoder(df[col])
  
```

```python
X = df.drop("G3", axis=1)
y = df["G3"]
```

```python
from sklearn.feature_selection import SelectKBest, f_regression
selector = SelectKBest(score_func=f_regression, k=20)
X_secilen = selector.fit_transform(X, y)
```

```python
#Özelliklerin indeksleri
secilen_indeksler = selector.get_support(indices=True)

#Orijinal özellik isimleri
secilen_ozellik_isimleri = X.columns[secilen_indeksler]

print("Seçilen Önemli 20 Özellik:")
print(secilen_ozellik_isimleri)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_secilen, y, test_size=0.2,random_state = 42)
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import r2_score

rf = RandomForestRegressor(n_estimators=100) #100 ağaç kullanılması belirlenmiş.
rf.fit(X_train, y_train)
```

```python
y_pred = rf.predict(X_test)
r2_random=r2_score(y_test, y_pred)
print("RandomForest Regression R^2 Score:", r2_random)
```

```python
plt.figure()
plt.scatter(y_test, y_pred)
plt.scatter(y_test, y_test)
plt.plot(y_test, y_test)
plt.legend(["Predicted", "Original", "Regression Line"])
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score

linear_reg = LinearRegression()
linear_reg.fit(X_train, y_train)
```

```python
y_pred_linear = linear_reg.predict(X_test)
r2_linear = r2_score(y_test, y_pred_linear)
print("Linear Regression R^2 Score:", r2_linear)
```

```python
# ORTALAMA KARESEL HATA HESAPLAMAK İÇİN

from sklearn.metrics import mean_squared_error
linear_model = LinearRegression()
linear_model.fit(X_train, y_train)

# Test verisi üzerinde tahmin yapma
linear_y_pred = linear_model.predict(X_test)

 #Ortalama karesel hata (Mean Squared Error) hesaplama
mse = mean_squared_error(y_test, linear_y_pred)
print(f"Linear Regression Mean Squared Error: {mse}")
```

```python
# Scatter plot ile gerçek ve tahmin edilen değerleri göster
plt.scatter(y_test, y_pred_linear, color='blue', label='Predicted')
plt.scatter(y_test, y_test, color='red', label='Original')
plt.plot(y_test, y_test, color='green', linewidth=2, label='Regression Line')

# Grafik üzerindeki elemanları düzenle
plt.title('Linear Regression: Predicted vs Original')
plt.xlabel('Original Values')
plt.ylabel('Predicted Values')
plt.legend()

# Grafik göster
plt.show()
```

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.datasets import make_regression
decision_tree_reg = DecisionTreeRegressor(random_state=42)
decision_tree_reg.fit(X_train, y_train)
```

```python
y_pred_dt = decision_tree_reg.predict(X_test)
r2_dt = r2_score(y_test, y_pred_dt)
print("Decision Tree Regression R^2 Score:", r2_dt)
```

```python
plt.scatter(y_test, y_pred_dt, color='blue', label='Predicted')
plt.scatter(y_test, y_test, color='red', label='Original')
plt.plot(y_test, y_test, color='green', linewidth=2, label='Regression Line')
plt.title('Decision Tree Regression: Predicted vs Original')
plt.xlabel('Original Values')
plt.ylabel('Predicted Values')
plt.legend()
plt.show()
```

```python
plt.bar(x=['RandomForestRegressor','LinearRegression','DecisionTreeRegressor'],height=[r2_random,r2_linear ,r2_dt])
plt.xlabel("Algorithms")
plt.ylabel("R^2 Score")
plt.show()
```
